```python
df['E-C'] = df['期望報酬率'] - df['累積報酬率']
df['E-C'].describe()
```

```python
import numpy as np
from math import erf, sqrt
mean = df['E-C'].mean()
sigma = df['E-C'].std()
prob = 1-(0.5*(1 + erf((0-mean) / (sigma*sqrt(2)))))
```
